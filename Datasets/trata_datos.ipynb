`Hacemos la asignacion de variable y las transformaciones para sustituir los valores, generar campo id con el formato requerido (Amazon = as123), utilizando y aplicando basicamente el metodo ".astype('str')" a las columnas que realmente precisamos`

# Limpieza de Datos

## Generar campo **`id`**: id se compondrá de la primera letra del nombre de la plataforma en cuestión, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = **`as123`**)

```python
amazon = pd.read_csv(r"amazon_prime_titles-score.csv")
disney = pd.read_csv(r"disney_plus_titles-score.csv")
hulu = pd.read_csv(r"hulu_titles-score (2).csv")
netflix = pd.read_csv(r"netflix_titles-score.csv")
```

```python
# amazon
amazon['show_id'] = amazon['show_id'].astype('str')
amazon['type'] = amazon['type'].astype('category')
amazon['title'] = amazon['title'].astype('str')
amazon['director'] = amazon['director'].astype('str')
amazon['cast'] = amazon['cast'].astype('str')
amazon['country'] = amazon['country'].astype('str')
amazon['date_added'] = pd.to_datetime(amazon['date_added'], infer_datetime_format=True, errors='coerce')
amazon['release_year'] = amazon['release_year'].astype('int16')
amazon['rating'] = amazon['rating'].astype('str')
amazon['listed_in'] = amazon['listed_in'].astype('str')
amazon['description'] = amazon['description'].astype('str')
amazon['score'] = amazon['score'].astype('int8')
```

```python
amazon['id'] = 'a' + amazon['show_id']
```

```python
amazon['id'] = amazon['id'].astype('str')
```
